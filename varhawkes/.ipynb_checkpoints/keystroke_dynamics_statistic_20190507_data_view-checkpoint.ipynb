分析击键行为数据的概率分布。Killourhy和Maxion的研究表明，keystroke dynamics的精度大概是16ms.这里把他的数据，按此精度划分。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
from scipy.stats import norm #使用直方图和最大似然高斯分布拟合绘制分布
import scipy
from scipy import stats
import scipy.stats.morestats as morestates
#不显示warning，为了好看用。需要显示warning时，注释下面两行
import warnings
warnings.filterwarnings('ignore')

#读取数据
df = pd.read_csv('../../data/DSL-StrongPasswordData.csv')

#对用户进行label编码，即从0编到50.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['target'] = le.fit_transform(df['subject'])
df.drop(['subject'], axis=1, inplace=True)

#拆分数据
ud_features = [column for column in df.columns if 'UD' in column]
dd_features = [column for column in df.columns if 'DD' in column]
hold_features = [column for column in df.columns if 'H' in column]

#df.head()
#df[dd_features].describe()
df[hold_features].describe()
```

```python
frames = [df[dd_features]['DD.period.t'], df[dd_features]['DD.t.i'], df[dd_features]['DD.i.e'], df[dd_features]['DD.e.five'], df[dd_features]['DD.five.Shift.r'],df[dd_features]['DD.Shift.r.o'], df[dd_features]['DD.o.a'], df[dd_features]['DD.a.n'],df[dd_features]['DD.n.l'], df[dd_features]['DD.l.Return']]
interval_DD = pd.concat(frames)
interval_DD.describe()
#print(np.log(interval_DD))
#interval_DD = np.log(interval_DD)
```

```python
sns.kdeplot(np.log(interval_DD),cut=0)
```

```python
sns.kdeplot(np.log(interval_DD),cumulative=True)
```

```python
sns.distplot(np.log(interval_DD),rug=True,rug_kws={"color": "g"},
             kde_kws={"color": "r", "lw": 1, "label": "KDE"},
             hist_kws={"histtype": "step", "linewidth": 1,"alpha": 1, "color": "b"},
             fit = stats.loglaplace, norm_hist = True )

#fit = stats.gamma
#fit = stats.norm
#fit = stats.expon
#fit = stats.alpha
#fit = stats.beta
#fit = stats.skewnorm
#fit = stats.powernorm
#fit = stats.logistic
#fit = stats.loggamma
#fit = stats.loglaplace
```

```python
#1.所有键所有人的DD
# -*- coding: utf-8 -*-

A2,critical,sig = morestates.anderson(np.log(interval_DD), dist = 'norm')
#dist : {'norm','expon','logistic','gumbel','gumbel_l', gumbel_r','extreme1'}
print(A2)
print(critical)
print(sig)
#结果表明不符合正态分布
#410.8116658818617
#[0.576 0.656 0.787 0.918 1.092]
#[15.  10.   5.   2.5  1. ]
```

```python
A2,critical,sig = morestates.anderson(np.log(interval_DD), dist = 'expon')
#dist : {'norm','expon','logistic','gumbel','gumbel_l', gumbel_r','extreme1'}
print(A2)
print(critical)
print(sig)
#结果表明不符合指数分布
#inf
#[0.922 1.078 1.341 1.606 1.957]
#[15.  10.   5.   2.5  1. ]
```

```python
A2,critical,sig = morestates.anderson(np.log(interval_DD), dist = 'logistic')
#dist : {'norm','expon','logistic','gumbel','gumbel_l', gumbel_r','extreme1'}
print(A2)
print(critical)
print(sig)
#结果表明也不符合logistic分布
#397.7863393406151
#[0.426 0.563 0.66  0.769 0.906 1.01 ]
#[25.  10.   5.   2.5  1.   0.5]
```

```python
A2,critical,sig = morestates.anderson(np.log(interval_DD), dist = 'gumbel')
#dist : {'norm','expon','logistic','gumbel','gumbel_l', gumbel_r','extreme1'}
print(A2)
print(critical)
print(sig)
#结果表明也不符合gumbel分布
#5010.384940284304
#[0.474 0.637 0.757 0.877 1.038]
#[25.  10.   5.   2.5  1. ]
```

```python
A2,critical,sig = morestates.anderson(np.log(interval_DD), dist = 'extreme1')
#dist : {'norm','expon','logistic','gumbel','gumbel_l', gumbel_r','extreme1'}
print(A2)
print(critical)
print(sig)
#结果表明也不符合extreme1分布
#5010.384940284304
#[0.474 0.637 0.757 0.877 1.038]
#[25.  10.   5.   2.5  1. ]
```
