```python
import matplotlib.pyplot as plt

from tick.hawkes import SimuHawkesSumExpKernels, HawkesSumExpKern
from tick.plot import plot_point_process
from tick.plot import plot_hawkes_kernel_norms

from make_data import make_data
from make_data_for_estimate import make_data_for_estimate

decays = [3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0, 3.0]
#baseline = [0.12, 0.07]
#adjacency = [[[0, .1, .4], [.2, 0., .2]], [[0, 0, 0], [.6, .3, 0]]]

fileName = "../data/DSL-StrongPasswordData.xls"
keystroke_data = make_data('s002',0,40,fileName)

learner = HawkesSumExpKern(decays, penalty='elasticnet', elastic_net_ratio=0.8)
learner.fit(keystroke_data)

plot_hawkes_kernel_norms(learner)

keystroke_data_test = make_data_for_estimate('s002',200,fileName)

t_min = 0.000001
t_max = 6
fig, ax_list = plt.subplots(11, 1, figsize=(10, 25))
learner.plot_estimated_intensity(keystroke_data_test, t_min=t_min,t_max=t_max, ax=ax_list)
```
